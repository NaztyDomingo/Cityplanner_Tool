```python
import xlsx_to_csv_converter as xtcc
import filehandler_helper as fh
import transform as t
import pandas as pd

xtcc.convert_single_file('sweden_data', 'sweden_data', 'countyreport_CO2')

filename = 'countyreport_CO2.csv'
folder = 'sweden_data'
filepath = fh.get_path_of_file(folder, filename)
data = pd.read_csv(filepath, index_col=False)

```

```python
data = t.translate_replace(data)
```

```python
data = data[(data['Region'] != 'Alla') & (data['City'] == 'Alla')]

data
```

```python
data = t.drop_columns(data)

data
```

```python
data_melted = data.melt(id_vars=['Main sector', 'Region', 'City'], 
                            var_name='Year', 
                            value_name='Emission')

```

```python
data_melted['Year'] = data_melted['Year'].str.extract(r'(\d{4})').astype(int)
```

```python
data_melted['Emission'] = data_melted['Emission'] / 1000
```

```python
data_pivoted = data_melted.pivot_table(index=['Year', 'City', 'Region'], 
                                           columns='Main sector', 
                                           values='Emission', 
                                           aggfunc='sum').reset_index()
```

```python
for sector in ['Transportation', 'Industry', 'Agriculture', 'Electricity And District Heating', 
                   'Other Heating', 'Machinery', 'Waste And Sewage']:
        if sector not in data_pivoted:
            data_pivoted[sector] = 0
```

```python
data_pivoted['Total Emissions'] = data_pivoted[['Waste And Sewage', 'Machinery', 'Electricity And District Heating', 
                                                   'Other Heating', 'Agriculture', 'Transportation', 'Industry']].sum(axis=1)
```

```python
final_columns = ['Year', 'City', 'Region', 'Waste And Sewage', 'Machinery', 'Electricity And District Heating', 
                     'Other Heating', 'Agriculture', 'Transportation', 'Industry', 'Total Emissions']
data_pivoted = data_pivoted[final_columns]

data_pivoted = data_pivoted.drop(columns=['Main sector'])
data_pivoted = data_pivoted.reset_index(drop=True)
```
