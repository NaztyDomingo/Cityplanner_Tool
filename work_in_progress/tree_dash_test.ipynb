```python
import pandas as pd
import glob
import os

CURR_DIR_PATH = os.getcwd()

file_path = glob.glob(os.path.join(CURR_DIR_PATH, 'sweden_cities_emission.csv'))[0]
output_file = os.path.join(CURR_DIR_PATH, 'sweden_cities_emissions.csv')

def read_file(file_path: str) -> pd.DataFrame:
    data = pd.read_csv(file_path)
    return data

def rename_reorder_columns(data: pd.DataFrame) -> pd.DataFrame:
    data = data.rename(columns={'Total Emission': 'Total Emissions'})
    return data

def data_to_csv(data: pd.DataFrame, output_file: str) -> None:
    data.to_csv(output_file, index=False)

# Read the data
data = read_file(file_path)

# Rename and reorder columns
data = rename_reorder_columns(data)

# Write the data to CSV
data_to_csv(data, output_file)

```
